```python
df_hand = pd.read_csv("./finger_tapping/finger_tapping_dataset_non_manifest_control.csv")
# if Participant_ID  starts with '20', remove first 20 chars
def parse_patient_id(name:str):
        if name.startswith("NIH"): [ID, *_] = name.split("-")
        elif name.endswith("finger_tapping.mp4"): [*_, ID, _] = name.split("-")
        else: [*_, ID, _, _, _] = name.split("_")
        return ID
df_hand['Participant_ID'] = df_hand.filename.apply(parse_patient_id)
df_hand.to_csv("./finger_tapping/finger_tapping_dataset_non_manifest_control_parsed.csv", index=False)
df_hand.shape, df_hand['Participant_ID'].nunique()
```

```python
df_hand_carrier = df_hand[df_hand['Participant_ID'].isin(fusion_ids['id'])]
df_hand_carrier.shape, df_hand_carrier['Participant_ID'].nunique()
```

```python
df_hand_carrier.to_csv("./finger_tapping/finger_tapping_dataset_non_manifest_control_updated.csv", index=False)
```
